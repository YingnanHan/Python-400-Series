```python
import pandas as pd
```

#### 读取Excel文件

```python
data = pd.read_excel("excelTest.xlsx")#按照转义路径读取文件
data
```

```python
data = pd.read_excel(r"excelTest.xlsx")#按照原路径不进行转义读取文件
data
```

```python
data = pd.read_excel(r"excelTest.xlsx",sheet_name="Sheet1")#读取指定名称的Sheet中的内容
data
```

```python
data = pd.read_excel(r"excelTest.xlsx",sheet_name=0)#按照序(Sheet索引)号读取Sheet中的内容
data
```

#### 指定固定的列作为行索引

```python
data = pd.read_excel(r"excelTest.xlsx",index_col=0)
data
```

```python
data = pd.read_excel(r"excelTest.xlsx",index_col=2)
data
```

#### 指定固定的列作为行索引

```python
data = pd.read_excel("excelTest.xlsx",header=0)#默认指定第一行作为列索引
data
```

```python
data = pd.read_excel("excelTest.xlsx",header=3)#指定第三行作为列索引
data
```

#### 使用默认的列索引

```python
data = pd.read_excel("excelTest.xlsx",header=None)
data
```

#### 指定导入列

```python
#需求：只看姓名与年龄列
data = pd.read_excel("excelTest.xlsx",usecols=[2])#仅仅传数字的方法已经不再被支持，想要显示那几列，就将列号传入
data
```
